# Queue Implementation

```python
class Queue:
    def __init__(self):
        self.qu = []
    def isempty(self):
        return self.qu == []
    def add(self,element):
        self.qu.append(element)
    def display(self):
        return self.qu
    def delete(self):
        if self.isempty():
            return -1
        else:
            return self.qu.pop(0)
q = Queue()
ch = 0
while ch<3:
    print('1: Addition ')
    print('2: Delete ')
    print('3: Exit ')
    ch = int(input('Enter your choice '))
    if(ch == 1):
       element = float(input('Enter element '))
       q.add(element)
    elif ch == 2:
        element = q.delete()
        if element == -1:
            print('Queue is empty...')
        else:
            print('Removed element ',element)
    else:
         break
# Print the queue
    print('queue =',q.display())
```

```python
class Queue:
    def __init__(self):
        self.qu = []
    def isempty(self):
        return self.qu == []
    def add(self,element):
        self.qu.append(element)
    def display(self):
        return self.qu
    def delete(self):
        if self.isempty():
            return -1
        else:
            return self.qu.pop(0)
    def search(self, element):
        if self.isempty():
             return -1
        else:
            try:
                n= self.qu.index(element)
                return len(self.qu)-n
            except ValueError:
                return -2
q = Queue()
ch = 0
while ch<3:
    print('1: Addition ')
    print('2: Delete ')
    print('3: search for element')
    print('4: Exit ')
    ch = int(input('Enter your choice '))
    if(ch == 1):
       element = float(input('Enter element '))
       q.add(element)
    elif ch == 2:
        element = q.delete()
        if element == -1:
            print('Queue is empty...')
        else:
            print('Removed element ',element)
    elif ch == 3:
        element = int(input('enter element: '))
        pos = q.search(element)
        if pos ==-1:
            print('the stack is empty')
        elif pos==-2:
            print('element not found in the stack')
        else:
             print('element found at position: ',pos)
    else:
         break
# Print the queue
    print('queue =',q.display())
```
