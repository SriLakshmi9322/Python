## Handling Missing Data

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\HP\Desktop\pandas\Book5.csv')
df
```

```python
# Using fillna()
df1 = df.fillna(0)
df1
```

```python
df1=df.fillna({'ename':'Name Missing','salary':0.0,'doj':'00-00-0000'})
df1
```

```python
# To remove those rows having Na or NaN use dropna()
df1=df.dropna()
df1
```

```python
df1=pd.DataFrame({'Player Name':['Rahul Dravid','Virat','Vinod Kamble'],
                   'Age':[30,28,22]})
df1
```

```python
df2=pd.DataFrame({'Player Name':['Rahul Dravid','Virat','Harbajan'],
                   'Salary':[45,60,44]})
df2
```

```python
# To join two data frames, we use merge() function
pd.merge(df1,df2,on='Player Name',how='inner')
```

```python
# Left join
pd.merge(df1,df2,on='Player Name',how='left')
```

```python
# Right join
pd.merge(df1,df2,on='Player Name',how='right')
```

```python
# Outer join
pd.merge(df1,df2,on='Player Name',how='outer')
```

```python
df3=pd.DataFrame({'Sports Person':['Rahul Dravid','Virat','Harbajan'],
                   'Salary':[45,60,44]})
df3
```

### How to join when there is no Common Column

```python
pd.merge(df1,df3,left_on='Player Name',right_on='Sports Person')
```

```python
pd.merge(df1,df3,left_on='Player Name',right_on='Sports Person',how='left')
```

```python
pd.merge(df1,df3,left_on='Player Name',right_on='Sports Person',how='right')
```

```python
pd.merge(df1,df3,left_on='Player Name',right_on='Sports Person',how='outer')
```

```python
pd.concat([df1,df3])
```

```python
pd.concat([df1,df3],ignore_index=True)
```

```python
pd.concat([df1,df3],axis=1)
```

## Where() Method

```python
import pandas as pd
```

```python
students={'Name':['Vinay','Anil','Gopal','Nani'],
          'Test1':[15,45,60,76],
          'Test2':[45,56,78,80],
          'Test3':[89,90,45,40]}
```

```python
df=pd.DataFrame(students)
df
```

```python
df.set_index('Name',inplace=True)
df
```

```python
df.where(df<60,'First')
```

```python
df.where(df['Test1']<60,'First')
```

```python
df.where(lambda x:x<80,'Excellent')
```

```python
df.where(lambda x:x+10>=30,'Poor')
```

## groupby() Method

```python
climate={'City':['Delhi','Delhi','Delhi','Delhi','Pune','Pune','Pune','Pune'],
         'Temp':[35,33,38,35,49,50,44,48],
         'Wind':[10,8,7,10,15,12,11,13]}
```

```python
df=pd.DataFrame(climate,columns=["City","Temp","Wind"])
df
```

```python
df1=df.groupby('City')
df1
```

```python
for city,rows in df1:
    print(city)
    print(rows)
```

```python
df1.get_group('Pune')
```

```python
df1.get_group('Pune').max()
```

```python
df1.max()
```

```python
df1.agg('max')
```

```python
df1.agg(['max','min','mean'])
```

```python
df1.agg(['max','count'])
```

```python
df['Temp'].agg(['max','min'])
```

```python
df1.agg({"Temp":['max','min'],
        'Wind':['mean','median']})
```
