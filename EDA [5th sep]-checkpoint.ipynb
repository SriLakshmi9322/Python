```python
cars= cars.drop(['MSRP','Invoice'],axis=1)
cars
```

```python
# To keep only one row and remove all its other duplicates.

cars = cars.drop_duplicates(keep='first')
cars
```

```python
# To know how many rows are having missing values in each column
cars.isnull().sum()
```

```python
# To delete missing value
cars.dropna(inplace=True)
cars
```

```python
# To sort the rows depending on a column
cars_sort = cars.sort_values(by='MPG_City', ascending=False)
cars_sort
```

```python
# Find out how many continents are manufacturing the cars

cars['Origin'].unique()
```
